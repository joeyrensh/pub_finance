```python
import pandas as pd
df = pd.read_csv('./position_log.txt', usecols=[i for i in range(1, 7)])
df.sort_values(by=['buy_date', 'p&l'], ascending=False, inplace=True)
df.reset_index(drop=True, inplace=True)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib
```

```python
cm = sns.color_palette("Blues", as_cmap=True)
df.style.set_table_styles([{"selector": "thead","props": "background-color:purple;color:white;"}]).format({"price": "{:.2f}","adjbase": "{:.2f}","p&l": "{:.2f}"}).background_gradient(subset=['price','adjbase'],cmap=cm).bar(subset=['p&l'],align='mid',color=['#5fba7d', '#d65f5f']).hide_index().set_properties(subset=['price', 'adjbase', 'p&l'], **{'width': '15px'})

```
